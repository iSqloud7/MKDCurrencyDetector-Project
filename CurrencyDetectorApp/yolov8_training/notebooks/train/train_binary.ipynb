```python
from ultralytics import YOLO
import os

CURRENT_DIR = os.getcwd() # yolov8_training/notebooks/train
print("Current directory: {}".format(CURRENT_DIR))
BASE_DIR = os.path.abspath(os.path.join(CURRENT_DIR, "..", "..")) # yolov8_training
print("BASE_DIR: {}".format(BASE_DIR))

DATA_YAML = os.path.join(BASE_DIR, "datasets", "binary", "data.yaml") # yolov8_training/datasets/binary/data.yaml
print("DATA_YAML: {}".format(DATA_YAML))
SAVE_DIR = os.path.join(BASE_DIR, "models", "binary_detector") # yolov8_training/datasets/models/binary_detector
print("SAVE_DIR: {}".format(SAVE_DIR))

model = YOLO(os.path.join(BASE_DIR, "notebooks", "train", "yolov8s.pt"))

print(f"Using dataset: {DATA_YAML}")
print(f"Training results will be saved in: {SAVE_DIR}")

model.train(
    data=DATA_YAML,
    epochs=100,
    imgsz=640,
    batch=8,
    project=SAVE_DIR,
    name="weights",
    device=0
)
```
