```python
from ultralytics import YOLO
import os

MODEL_DIR = "../backend/app/models"
os.makedirs(MODEL_DIR, exist_ok=True)

models_to_export = {
    "binary": "../mkd_currency_model/binary_train/weights/best.pt",
    "banknote": "../mkd_currency_model/banknote_train/weights/best.pt",
    "coin": "../mkd_currency_model/coin_train/weights/best.pt"
}

for name, path in models_to_export.items():
    model = YOLO(path)
    model.export(format="torchscript", device="cuda")
    print(f"{name} model exported to TorchScript at {MODEL_DIR}")
```
