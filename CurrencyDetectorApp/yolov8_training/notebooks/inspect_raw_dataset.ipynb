```python
import os

import yaml

CURRENT_DIR = os.getcwd() # yolov8_training/notebooks
print("Current directory: {}".format(CURRENT_DIR))
BASE_DIR = os.path.dirname(CURRENT_DIR) # yolov8_training
print("Base directory: {}".format(BASE_DIR))
TARGET_DIR = os.path.join(BASE_DIR, "datasets", "raw_all_classes", "data.yaml") # yolov8_training/datasets/raw_all_classes/data.yaml
print("Target directory: {}".format(TARGET_DIR))

DATA_YAML = TARGET_DIR # yolov8_training/datasets/raw_all_classes/data.yaml
print("Data YAML: {}".format(DATA_YAML))

with open(DATA_YAML, "r", encoding="utf-8") as f:
    data = yaml.safe_load(f)

print("üìä Number of classes:", data["nc"])
print("üè∑Ô∏è Class names:")

for idx, name in enumerate(data["names"]):
    print(f"  {idx}: {name}")

print("\nüìÅ Dataset paths:")
print(f"  Train: {data.get('train', 'N/A')}")
print(f"  Valid: {data.get('val', 'N/A')}")
print(f"  Test: {data.get('test', 'N/A')}")
```
